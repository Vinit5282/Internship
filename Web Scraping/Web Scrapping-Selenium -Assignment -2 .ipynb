## Q-1 In this question you have to scrape data using the filters available on the webpage You have to use the location and
salary filter.
You have to scrape data for “Data Scientist” designation for first 10 job results.
You have to scrape the job-title, job-location, company name, experience required.
The location filter to be used is “Delhi/NCR”. The salary filter to be used is “3-6” lakhs
The task will be done as shown in the below steps:
1. first get the web page https://www.naukri.com/
2. Enter “Data Scientist” in “Skill, Designations, and Companies” field.
3. Then click the search button.
4. Then apply the location filter and salary filter by checking the respective boxes
5. Then scrape the data for the first 10 jobs results you get.
6. Finally create a dataframe of the scraped data.

```python
# Opening the web page 
driver.get("https://www.naukri.com/")
```

```python
designation= driver.find_element(By.CLASS_NAME,"suggestor-input ")
designation.send_keys('Data Scientist')
```

```python
search = driver.find_element(By.CLASS_NAME,"qsbSubmit")
search.click()
```

```python
job_title = []
job_location =[]
company_name =[]
experience_required =[]
```

```python
title_tags = driver.find_elements(By.XPATH,'//div[@class="cust-job-tuple layout-wrapper lay-2 sjw__tuple "]/div/a')
for i in title_tags:
    title = i.text 
    job_title.append(title)
    
    
location_tags = driver.find_elements(By.XPATH,'//span[@class="ni-job-tuple-icon ni-job-tuple-icon-srp-location loc"]/span')
for i in location_tags:
    location = i.text
    job_location.append(location)
    
    
company_tags = driver.find_elements(By.XPATH,'//span[@class=" comp-dtls-wrap"]/a[1]')
for i in company_tags:
    company = i.text
    company_name.append(company)
    
    
experience_tags = driver.find_elements(By.XPATH,'//span[@class="ni-job-tuple-icon ni-job-tuple-icon-srp-experience exp"]/span')
for i in experience_tags:
    experience = i.text
    experience_required.append(experience)
```

```python
print(len(job_title),len(job_location),len(company_name),len(experience_required))
```

```python
import pandas as pd 
df = pd.DataFrame({'Title':job_title,'Location':job_location,'Company_name':company_name,'Experience':experience_required})
df[0:10]
```

## Q6: Write a python program to scrape data for Top 1000 Quotes of All Time.
The above task will be done in following steps:
1. First get the webpagehttps://www.azquotes.com/
2. Click on Top Quote
3. Than scrap a) Quote b) Author c) Type Of Quotes

```python
driver = webdriver.Chrome()
```

```python
# Opening the web page 
driver.get(" https://www.azquotes.com/")
```

```python
top_quote = driver.find_element(By.XPATH,'/html/body/div[1]/div[1]/div[1]/div/div[3]/ul/li[5]/a')
top_quote.click()
```

```python
quotes = []
authors =[]
type_of_quotes = []
```

```python
quote = driver.find_elements(By.XPATH,'//a[@class="title"]')
for i in quote:
    quote = i.text 
    quotes.append(quote)
    
    
author = driver.find_elements(By.XPATH,'//div[@class="author"]/a')
for i in author:
    author = i.text
    authors.append(author)
    
    
type_of_quote = driver.find_elements(By.XPATH,'//div[@class="tags"]')
for i in type_of_quote:
    type_of_quote = i.text
    type_of_quotes.append(type_of_quote)
```

```python
print(len(quotes),len(authors),len(type_of_quotes))
```

```python
df = pd.DataFrame({'Quotes':quotes,'Authors':authors,'Type_of_Quotes':type_of_quotes})
df
```

```python
Top1000_Quote = []
```

```python
start = 0
end = 10 
for page in range (start,end):
    Top1000 = driver.find_elements(By.XPATH,'//a[@class="title"]')
    for i in Top1000:
        Top1000 = i.text 
        Top1000_Quote.append(Top1000)
  
    
    next_button= driver.find_element(By.XPATH,'/html/body/div[1]/div[2]/div/div/div/div[1]/div/div[4]/li[12]/a')
    next_button.click()
    time.sleep(3)
```

```python
Top1000 = pd.DataFrame({'Top1000_Quotes':Top1000_Quote})
Top1000
```
