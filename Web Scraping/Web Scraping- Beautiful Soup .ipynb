# Question-3 Write a python program to scrape mentioned details from dineout.co.in : i) Restaurant name
ii) Cuisine iii) Location iv) Ratings v) Image URL.

```python

page3 = requests.get('https://www.dineout.co.in/delhi-restaurants/welcome-back')
page3
```

```python
soup = BeautifulSoup(page3.content)
soup
```

```python
import pandas as pd
cusine = []
for i in soup.find_all('span', class_="double-line-ellipsis"):
    cuisine_text = i.text.split('|')[-1].strip()
    cusine.append(cuisine_text)

df = pd.DataFrame(cusine, columns=['Cuisine'])
print(df)

```

```python
location = []

for i in soup.find_all('div',class_="restnt-loc ellipsis"):
    location.append(i.text)
location
```

```python
rating = []

for i in soup.find_all('div',class_="restnt-rating rating-4"):
    rating.append(i.text)
rating
```

```python
img_url = []

for i in soup.find_all('img',class_="no-img"):
    img_url.append(i['data-src'])
img_url
```

```python

import pandas as pd


cusine_series= pd.Series(cusine).fillna('NA')  # Fill missing value 
location_series = pd.Series(location).fillna('NA')
rating_series = pd.Series(rating).fillna('NA')
img_url_series = pd.Series(img_url).fillna('NA')
df = pd.DataFrame({"Cuisine": cusine_series, "Location": location_series, "Rating":rating_series,"Img_URL":img_url_series})

df
```

# Question-4 Write s python program to display list of respected former finance minister of India(i.e.
Name , Term of office) from https://presidentofindia.nic.in/former-presidents.htm and make
data frame.

```python

page4 = requests.get('https://presidentofindia.nic.in/former-presidents')
page4
```

```python
soup = BeautifulSoup(page4.content)
soup
```

```python

import re

Name_of_President = []

for i in soup.find_all('div', class_="desc-sec"):
    president_name = i.text.replace('\n', '')
    if "5th" not in president_name and "6th" not in president_name:
        Name_of_President.append(president_name)

presidents_names = [re.split(r'\d+(st|nd|rd|th)', name)[0].strip() for name in Name_of_President]

print(presidents_names)
```

```python


Name_of_President = []

for i in soup.find_all('div', class_="desc-sec"):
    president_name = i.text.replace('\n', '')
    if "5th" not in president_name and "6th" not in president_name:
        Name_of_President.append(president_name)

presidents_info = []

for name in Name_of_President:
    match = re.match(r'(.+?)(\d+(?:st|nd|rd|th))(.+)', name)
    if match:
        ordinal_number = match.group(2)
        later_part = match.group(3)
        presidents_info.append(ordinal_number + later_part.strip())

print(presidents_info)
```

```python
import pandas as pd


President_Name= pd.Series(presidents_names).fillna('NA')  # Fill missing value 
Terms_of_office = pd.Series(presidents_info).fillna('NA')

df = pd.DataFrame({"Name_of_Presidents":presidents_names , "Term_of_office": Terms_of_office})

df
```
